```python
%load_ext raw_magic
```

```python
%%view movies
read("https://raw.githubusercontent.com/prust/wikipedia-movie-data/master/movies.json", cache := interval "1 day")
```

```python
%%query
SELECT *
FROM movies
```

```python
%%query
SELECT *
FROM movies
WHERE ccount(`cast`) > 0
```

```python
%%view m2
SELECT title, year, genres,
    `cast` AS casting
FROM movies
WHERE ccount(`cast`) > 0
```

```python
%%query
SELECT *
FROM m2
```

```python
%%query
SELECT year, count(*)
FROM m2
GROUP BY year
ORDER BY count(*) DESC
```

```python
%%query
SELECT *
FROM m2
WHERE year = 2006
```

```python
%%query
SELECT name, (SELECT title FROM *) as movies
FROM m2 m, name IN m.casting
WHERE m.year = 2006 and name != "Jr."
GROUP BY name
ORDER BY count(*) DESC
```
